```python
friends = nx.Graph([("Diane","Carol"),("Diane","Andre"),("Diane","Fernando"),
                   ("Diane","Beverly"),("Diane","Garth"),
                    ("Diane","Ed"),
                    ("Ed", "Beverly"), ("Ed", "Garth"),
                    ("Andre", "Beverly"),("Andre", "Fernando"),
                    ("Garth", "Fernando"),("Garth", "Beverly"),
                    ("Garth","Heather"),
                    ("Fernando","Heather"),
                    ("Carol", "Fernando"), ("Carol","Andre"),
                    ("Heather","Ike"),
                    ("Ike", "Jane")
                   ])
```
